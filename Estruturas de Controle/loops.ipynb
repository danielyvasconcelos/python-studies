Loops em Python

2.1 Loop For Básico

```python
# For com listas
frutas = ['maçã', 'banana', 'laranja']
for fruta in frutas:
    print(f"Eu gosto de {fruta}")

# For com strings
palavra = "Python"
for letra in palavra:
    print(f"Letra: {letra}")

# For com range()
print("\nContando de 0 a 4:")
for i in range(5):
    print(f"Número: {i}")

print("\nContando de 1 a 10:")
for i in range(1, 11):
    print(f"Número: {i}")

print("\nContando de 2 em 2:")
for i in range(0, 11, 2):
    print(f"Par: {i}")

# For com dicionários
pessoa = {'nome': 'Ana', 'idade': 25, 'cidade': 'São Paulo'}

print("\nChaves:")
for chave in pessoa:
    print(chave)

print("\nValores:")
for valor in pessoa.values():
    print(valor)

print("\nChave-Valor:")
for chave, valor in pessoa.items():
    print(f"{chave}: {valor}")
```

2.2 Loop While

```python
# While básico
contador = 0
while contador < 5:
    print(f"Contador: {contador}")
    contador += 1

# While com condição de entrada do usuário (simulado)
def simular_menu():
    opcoes = ['1', '2', '3', 'sair']
    opcao_atual = 0
    
    while opcao_atual < len(opcoes) - 1:
        opcao = opcoes[opcao_atual]
        print(f"Opção selecionada: {opcao}")
        
        if opcao == '1':
            print("Executando opção 1")
        elif opcao == '2':
            print("Executando opção 2")
        elif opcao == '3':
            print("Executando opção 3")
        
        opcao_atual += 1
    
    print("Saindo do menu...")

simular_menu()

# While para validação
def validar_numero():
    numeros_teste = ['abc', '-5', '0', '10']  # Simulando entradas
    indice = 0
    
    while indice < len(numeros_teste):
        entrada = numeros_teste[indice]
        print(f"\nTestando entrada: '{entrada}'")
        
        if entrada.isdigit() and int(entrada) > 0:
            numero = int(entrada)
            print(f"Número válido: {numero}")
            break
        else:
            print("Número inválido. Deve ser um inteiro positivo.")
        
        indice += 1
    else:
        print("Nenhum número válido encontrado.")

validar_numero()

# While infinito com break
def processar_dados():
    dados = [10, 20, -1, 30, 40]  # -1 é o sinal de parada
    indice = 0
    soma = 0
    
    while True:
        if indice >= len(dados):
            print("Fim dos dados")
            break
        
        valor = dados[indice]
        
        if valor == -1:
            print("Sinal de parada encontrado")
            break
        
        soma += valor
        print(f"Processando: {valor}, Soma atual: {soma}")
        indice += 1
    
    return soma

resultado = processar_dados()
print(f"Soma final: {resultado}")
```

2.3 Break e Continue

```python
# Break - interrompe o loop
print("Exemplo com break:")
for i in range(10):
    if i == 5:
        print(f"Parando no {i}")
        break
    print(f"Número: {i}")

# Continue - pula para próxima iteração
print("\nExemplo com continue (só números pares):")
for i in range(10):
    if i % 2 != 0:  # Se for ímpar
        continue    # Pula para próxima iteração
    print(f"Par: {i}")

# Busca em lista com break
def buscar_usuario(usuarios, nome_procurado):
    for usuario in usuarios:
        if usuario['nome'] == nome_procurado:
            print(f"Usuário encontrado: {usuario}")
            return usuario
    
    print(f"Usuário '{nome_procurado}' não encontrado")
    return None

usuarios = [
    {'nome': 'Ana', 'idade': 25},
    {'nome': 'Bruno', 'idade': 30},
    {'nome': 'Carlos', 'idade': 35}
]

buscar_usuario(usuarios, 'Bruno')
buscar_usuario(usuarios, 'Diana')

# Processamento com continue
def processar_numeros(numeros):
    resultado = []
    
    for num in numeros:
        # Pular números negativos
        if num < 0:
            print(f"Pulando número negativo: {num}")
            continue
        
        # Pular zero
        if num == 0:
            print("Pulando zero")
            continue
        
        # Processar apenas números positivos
        quadrado = num ** 2
        resultado.append(quadrado)
        print(f"{num}² = {quadrado}")
    
    return resultado

numeros = [-2, 0, 3, -1, 4, 5]
quadrados = processar_numeros(numeros)
print(f"Quadrados: {quadrados}")

# Break em loops aninhados
def encontrar_par_soma(matriz, soma_alvo):
    for i, linha in enumerate(matriz):
        for j, valor in enumerate(linha):
            for k, linha2 in enumerate(matriz[i:], i):
                for l, valor2 in enumerate(linha2):
                    if i == k and j >= l:  # Evitar duplicatas
                        continue
                    if valor + valor2 == soma_alvo:
                        print(f"Par encontrado: {valor} + {valor2} = {soma_alvo}")
                        print(f"Posições: ({i},{j}) e ({k},{l})")
                        return (valor, valor2)
    
    print(f"Nenhum par soma {soma_alvo}")
    return None

matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
encontrar_par_soma(matriz, 10)
```

2.4 Else em Loops

```python
# Else em for - executa se o loop completar sem break
def verificar_numero_primo(n):
    if n < 2:
        return False
    
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            print(f"{n} não é primo (divisível por {i})")
            return False
    else:
        print(f"{n} é primo")
        return True

# Testando números
numeros_teste = [17, 18, 19, 20]
for num in numeros_teste:
    verificar_numero_primo(num)

# Else em while
def buscar_elemento(lista, elemento):
    indice = 0
    while indice < len(lista):
        if lista[indice] == elemento:
            print(f"Elemento '{elemento}' encontrado no índice {indice}")
            break
        indice += 1
    else:
        print(f"Elemento '{elemento}' não encontrado na lista")

lista = ['a', 'b', 'c', 'd']
buscar_elemento(lista, 'c')  # Encontrado
buscar_elemento(lista, 'z')  # Não encontrado

# Validação com else
def validar_senha(senha):
    criterios = [
        (len(senha) >= 8, "pelo menos 8 caracteres"),
        (any(c.isupper() for c in senha), "pelo menos uma maiúscula"),
        (any(c.islower() for c in senha), "pelo menos uma minúscula"),
        (any(c.isdigit() for c in senha), "pelo menos um número")
    ]
    
    for valido, descricao in criterios:
        if not valido:
            print(f"Senha inválida: falta {descricao}")
            return False
    else:
        print("Senha válida!")
        return True

senhas_teste = ['123', 'MinhaSenh@123', 'minhasenha', 'MINHASENHA']
for senha in senhas_teste:
    print(f"\nTestando: '{senha}'")
    validar_senha(senha)

# Processamento de arquivo com else
def processar_dados_ate_erro(dados):
    processados = []
    
    for item in dados:
        try:
            # Simular processamento que pode falhar
            if item == 'erro':
                raise ValueError(f"Erro ao processar: {item}")
            
            resultado = item.upper()
            processados.append(resultado)
            print(f"Processado: {item} -> {resultado}")
            
        except ValueError as e:
            print(f"Erro encontrado: {e}")
            break
    else:
        print("Todos os dados foram processados com sucesso!")
    
    return processados

# Teste com dados válidos
dados_validos = ['item1', 'item2', 'item3']
resultado1 = processar_dados_ate_erro(dados_validos)

print("\n" + "="*50)

# Teste com erro
dados_com_erro = ['item1', 'erro', 'item3']
resultado2 = processar_dados_ate_erro(dados_com_erro)
```

2.5 Enumerate e Zip

```python
# Enumerate - obtém índice e valor
frutas = ['maçã', 'banana', 'laranja']

print("Com enumerate:")
for indice, fruta in enumerate(frutas):
    print(f"{indice}: {fruta}")

print("\nCom enumerate começando do 1:")
for indice, fruta in enumerate(frutas, 1):
    print(f"{indice}. {fruta}")

# Zip - combina múltiplas listas
nomes = ['Ana', 'Bruno', 'Carlos']
idades = [25, 30, 35]
cidades = ['SP', 'RJ', 'BH']

print("\nCom zip:")
for nome, idade, cidade in zip(nomes, idades, cidades):
    print(f"{nome}, {idade} anos, mora em {cidade}")

# Enumerate + Zip para processamento avançado
def comparar_listas(lista1, lista2):
    diferencas = []
    
    for i, (val1, val2) in enumerate(zip(lista1, lista2)):
        if val1 != val2:
            diferencas.append({
                'indice': i,
                'valor1': val1,
                'valor2': val2
            })
    
    return diferencas

lista_a = [1, 2, 3, 4, 5]
lista_b = [1, 2, 9, 4, 7]

diferencas = comparar_listas(lista_a, lista_b)
print(f"\nDiferenças encontradas: {len(diferencas)}")
for diff in diferencas:
    print(f"Índice {diff['indice']}: {diff['valor1']} != {diff['valor2']}")

# Criando dicionário com zip
chaves = ['nome', 'idade', 'profissao']
valores = ['Maria', 28, 'Desenvolvedora']

pessoa = dict(zip(chaves, valores))
print(f"\nPessoa criada: {pessoa}")

# Processamento de dados tabulares
def processar_vendas_mensais():
    meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai']
    vendas = [1000, 1200, 800, 1500, 1100]
    metas = [1100, 1000, 900, 1400, 1200]
    
    relatorio = []
    
    for i, (mes, venda, meta) in enumerate(zip(meses, vendas, metas), 1):
        percentual = (venda / meta) * 100
        status = "✓" if venda >= meta else "✗"
        
        relatorio.append({
            'trimestre': (i-1) // 3 + 1,
            'mes': mes,
            'venda': venda,
            'meta': meta,
            'percentual': percentual,
            'atingiu_meta': venda >= meta,
            'status': status
        })
    
    return relatorio

relatorio = processar_vendas_mensais()
print("\nRelatório de Vendas:")
for item in relatorio:
    print(f"{item['status']} {item['mes']}: R$ {item['venda']} "
          f"(Meta: R$ {item['meta']}) - {item['percentual']:.1f}%")

# Zip com listas de tamanhos diferentes
from itertools import zip_longest

lista_curta = [1, 2, 3]
lista_longa = ['a', 'b', 'c', 'd', 'e']

print("\nZip normal (para no menor):")
for num, letra in zip(lista_curta, lista_longa):
    print(f"{num} - {letra}")

print("\nZip_longest (preenche com None):")
for num, letra in zip_longest(lista_curta, lista_longa, fillvalue='?'):
    print(f"{num} - {letra}")
```

2.6 Loops Aninhados

```python
# Matriz - loops aninhados básicos
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Percorrendo matriz:")
for i, linha in enumerate(matriz):
    for j, valor in enumerate(linha):
        print(f"matriz[{i}][{j}] = {valor}")

# Tabuada com loops aninhados
def gerar_tabuada(ate=5):
    print(f"\nTabuada de 1 a {ate}:")
    for i in range(1, ate + 1):
        print(f"\nTabuada do {i}:")
        for j in range(1, 11):
            resultado = i * j
            print(f"{i} x {j:2d} = {resultado:2d}")

gerar_tabuada(3)

# Busca em estrutura aninhada
def buscar_produto(categorias, produto_procurado):
    for categoria, produtos in categorias.items():
        for produto in produtos:
            if produto['nome'].lower() == produto_procurado.lower():
                print(f"Produto '{produto_procurado}' encontrado na categoria '{categoria}'")
                print(f"Detalhes: {produto}")
                return categoria, produto
    
    print(f"Produto '{produto_procurado}' não encontrado")
    return None, None

loja = {
    'eletrônicos': [
        {'nome': 'Notebook', 'preco': 2500},
        {'nome': 'Mouse', 'preco': 50}
    ],
    'livros': [
        {'nome': 'Python Cookbook', 'preco': 80},
        {'nome': 'Clean Code', 'preco': 90}
    ]
}

buscar_produto(loja, 'mouse')
buscar_produto(loja, 'tablet')

# Combinações com loops aninhados
def gerar_combinacoes(lista1, lista2):
    combinacoes = []
    
    for item1 in lista1:
        for item2 in lista2:
            combinacao = f"{item1}-{item2}"
            combinacoes.append(combinacao)
            print(f"Combinação: {combinacao}")
    
    return combinacoes

cores = ['azul', 'vermelho']
tamanhos = ['P', 'M', 'G']
combinacoes = gerar_combinacoes(cores, tamanhos)
print(f"\nTotal de combinações: {len(combinacoes)}")

# Processamento de dados hierárquicos
def calcular_vendas_por_regiao(dados_vendas):
    relatorio = {}
    
    for regiao, cidades in dados_vendas.items():
        total_regiao = 0
        relatorio[regiao] = {'cidades': {}, 'total': 0}
        
        for cidade, vendedores in cidades.items():
            total_cidade = 0
            
            for vendedor, vendas in vendedores.items():
                total_vendedor = sum(vendas)
                total_cidade += total_vendedor
                print(f"{regiao} > {cidade} > {vendedor}: R$ {total_vendedor}")
            
            relatorio[regiao]['cidades'][cidade] = total_cidade
            total_regiao += total_cidade
        
        relatorio[regiao]['total'] = total_regiao
    
    return relatorio

vendas = {
    'Sudeste': {
        'São Paulo': {
            'Ana': [1000, 1200, 800],
            'Bruno': [900, 1100, 1000]
        },
        'Rio de Janeiro': {
            'Carlos': [800, 900, 700]
        }
    },
    'Sul': {
        'Porto Alegre': {
            'Diana': [600, 700, 800]
        }
    }
}

relatorio = calcular_vendas_por_regiao(vendas)
print("\nResumo por região:")
for regiao, dados in relatorio.items():
    print(f"{regiao}: R$ {dados['total']}")
```

2.7 Otimização e Performance

```python
import time

# Comparação de performance: for vs while
def benchmark_loops(n=1000000):
    # For loop
    start = time.time()
    soma_for = 0
    for i in range(n):
        soma_for += i
    tempo_for = time.time() - start
    
    # While loop
    start = time.time()
    soma_while = 0
    i = 0
    while i < n:
        soma_while += i
        i += 1
    tempo_while = time.time() - start
    
    print(f"For loop: {tempo_for:.4f}s (soma: {soma_for})")
    print(f"While loop: {tempo_while:.4f}s (soma: {soma_while})")
    print(f"For é {tempo_while/tempo_for:.2f}x mais rápido")

benchmark_loops(100000)  # Reduzido para exemplo

# Evitar loops desnecessários
def busca_ineficiente(lista, item):
    """Busca ineficiente - sempre percorre toda a lista"""
    encontrado = False
    for elemento in lista:
        if elemento == item:
            encontrado = True
    return encontrado

def busca_eficiente(lista, item):
    """Busca eficiente - para quando encontra"""
    for elemento in lista:
        if elemento == item:
            return True
    return False

def busca_pythonica(lista, item):
    """Busca pythônica - usa 'in'"""
    return item in lista

# Teste de performance
lista_grande = list(range(10000))
item_procurado = 9999  # Último item

# Busca ineficiente
start = time.time()
resultado1 = busca_ineficiente(lista_grande, item_procurado)
tempo1 = time.time() - start

# Busca eficiente
start = time.time()
resultado2 = busca_eficiente(lista_grande, item_procurado)
tempo2 = time.time() - start

# Busca pythônica
start = time.time()
resultado3 = busca_pythonica(lista_grande, item_procurado)
tempo3 = time.time() - start

print(f"\nBusca ineficiente: {tempo1:.6f}s")
print(f"Busca eficiente: {tempo2:.6f}s")
print(f"Busca pythônica: {tempo3:.6f}s")

# Uso de generators para economia de memória
def numeros_pares_lista(n):
    """Retorna lista - usa mais memória"""
    return [i for i in range(n) if i % 2 == 0]

def numeros_pares_generator(n):
    """Retorna generator - usa menos memória"""
    for i in range(n):
        if i % 2 == 0:
            yield i

# Comparação de uso de memória
import sys

n = 1000
lista_pares = numeros_pares_lista(n)
gen_pares = numeros_pares_generator(n)

print(f"\nTamanho da lista: {sys.getsizeof(lista_pares)} bytes")
print(f"Tamanho do generator: {sys.getsizeof(gen_pares)} bytes")

# Processamento eficiente com break
def encontrar_primeiro_multiplo(numeros, divisor):
    """Encontra o primeiro múltiplo e para"""
    for i, num in enumerate(numeros):
        if num % divisor == 0:
            print(f"Primeiro múltiplo de {divisor}: {num} (posição {i})")
            return num
    
    print(f"Nenhum múltiplo de {divisor} encontrado")
    return None

numeros = [7, 13, 21, 35, 42, 49]
encontrar_primeiro_multiplo(numeros, 7)

# Dicas de otimização
print("\n=== DICAS DE OTIMIZAÇÃO ===")
print("✓ Use 'for' em vez de 'while' quando possível")
print("✓ Use 'break' para sair de loops quando encontrar o resultado")
print("✓ Use 'in' para verificar pertencimento em vez de loops")
print("✓ Considere generators para grandes volumes de dados")
print("✓ Evite loops aninhados desnecessários")
print("✓ Use enumerate() em vez de range(len())")
print("✓ Use zip() para iterar múltiplas sequências")
print("✓ Prefira list comprehensions para transformações simples")
```

2.8 Casos de Uso Práticos no Desenvolvimento

```python
# 1. Processamento de logs
def analisar_logs(logs):
    estatisticas = {
        'total_requests': 0,
        'status_codes': {},
        'ips_unicos': set(),
        'erros': []
    }
    
    for linha in logs:
        partes = linha.split(' ')
        if len(partes) < 4:
            continue
        
        ip = partes[0]
        status = partes[2]
        
        estatisticas['total_requests'] += 1
        estatisticas['ips_unicos'].add(ip)
        
        # Contar status codes
        if status in estatisticas['status_codes']:
            estatisticas['status_codes'][status] += 1
        else:
            estatisticas['status_codes'][status] = 1
        
        # Identificar erros
        if status.startswith('4') or status.startswith('5'):
            estatisticas['erros'].append(linha.strip())
    
    return estatisticas

logs_exemplo = [
    '192.168.1.1 GET 200 /home',
    '192.168.1.2 POST 404 /api/users',
    '192.168.1.1 GET 200 /about',
    '192.168.1.3 GET 500 /error'
]

stats = analisar_logs(logs_exemplo)
print("Estatísticas dos logs:")
print(f"Total de requests: {stats['total_requests']}")
print(f"IPs únicos: {len(stats['ips_unicos'])}")
print(f"Status codes: {stats['status_codes']}")
print(f"Erros encontrados: {len(stats['erros'])}")

# 2. Validação de dados em lote
def validar_usuarios_lote(usuarios):
    validos = []
    invalidos = []
    
    for i, usuario in enumerate(usuarios):
        erros = []
        
        # Validações
        if not usuario.get('nome'):
            erros.append('Nome obrigatório')
        
        email = usuario.get('email', '')
        if not email or '@' not in email:
            erros.append('Email inválido')
        
        idade = usuario.get('idade')
        if not isinstance(idade, int) or idade < 0:
            erros.append('Idade inválida')
        
        # Classificar resultado
        if erros:
            invalidos.append({
                'linha': i + 1,
                'usuario': usuario,
                'erros': erros
            })
        else:
            validos.append(usuario)
    
    return validos, invalidos

usuarios_teste = [
    {'nome': 'Ana', 'email': 'ana@email.com', 'idade': 25},
    {'nome': '', 'email': 'bruno@email.com', 'idade': 30},
    {'nome': 'Carlos', 'email': 'email_invalido', 'idade': -5}
]

validos, invalidos = validar_usuarios_lote(usuarios_teste)
print(f"\nUsuários válidos: {len(validos)}")
print(f"Usuários inválidos: {len(invalidos)}")

for item in invalidos:
    print(f"Linha {item['linha']}: {item['erros']}")

# 3. Processamento de dados financeiros
def calcular_metricas_vendas(vendas_mensais):
    metricas = {
        'total_ano': 0,
        'media_mensal': 0,
        'melhor_mes': {'mes': '', 'valor': 0},
        'pior_mes': {'mes': '', 'valor': float('inf')},
        'crescimento_mensal': [],
        'meses_acima_media': 0
    }
    
    # Primeira passada: calcular total e encontrar extremos
    for mes, valor in vendas_mensais.items():
        metricas['total_ano'] += valor
        
        if valor > metricas['melhor_mes']['valor']:
            metricas['melhor_mes'] = {'mes': mes, 'valor': valor}
        
        if valor < metricas['pior_mes']['valor']:
            metricas['pior_mes'] = {'mes': mes, 'valor': valor}
    
    # Calcular média
    metricas['media_mensal'] = metricas['total_ano'] / len(vendas_mensais)
    
    # Segunda passada: calcular crescimento e contar acima da média
    meses_ordenados = list(vendas_mensais.keys())
    for i, mes in enumerate(meses_ordenados):
        valor_atual = vendas_mensais[mes]
        
        # Contar meses acima da média
        if valor_atual > metricas['media_mensal']:
            metricas['meses_acima_media'] += 1
        
        # Calcular crescimento mensal
        if i > 0:
            mes_anterior = meses_ordenados[i-1]
            valor_anterior = vendas_mensais[mes_anterior]
            crescimento = ((valor_atual - valor_anterior) / valor_anterior) * 100
            metricas['crescimento_mensal'].append({
                'mes': mes,
                'crescimento': crescimento
            })
    
    return metricas

vendas_2023 = {
    'Jan': 10000, 'Fev': 12000, 'Mar': 11000,
    'Abr': 13000, 'Mai': 15000, 'Jun': 14000,
    'Jul': 16000, 'Ago': 18000, 'Set': 17000,
    'Out': 19000, 'Nov': 21000, 'Dez': 20000
}

metricas = calcular_metricas_vendas(vendas_2023)
print(f"\nMétricas de Vendas 2023:")
print(f"Total do ano: R$ {metricas['total_ano']:,}")
print(f"Média mensal: R$ {metricas['media_mensal']:,.0f}")
print(f"Melhor mês: {metricas['melhor_mes']['mes']} (R$ {metricas['melhor_mes']['valor']:,})")
print(f"Pior mês: {metricas['pior_mes']['mes']} (R$ {metricas['pior_mes']['valor']:,})")
print(f"Meses acima da média: {metricas['meses_acima_media']}/12")

print("\nCrescimento mensal:")
for item in metricas['crescimento_mensal'][-3:]:  # Últimos 3 meses
    print(f"{item['mes']}: {item['crescimento']:+.1f}%")
```
